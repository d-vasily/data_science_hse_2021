## Сортироовка пузырьком

```python
def bubble_sort(nums):
    for i in range(len(nums)):
        for j in range(len(nums) - i - 1):
            if nums[j] > nums[j + 1]:
                nums[j], nums[j + 1] = nums[j + 1], nums[j]

random_list_of_nums = [9, 5, 2, 1, 8, 0, 4, 4, 3, 7, 6]
bubble_sort(random_list_of_nums)
print(random_list_of_nums)
```

## Сортироовка выбором

```python
# Ищем минимальное значение и помещаем на место i, после увеличиваем i на 1.
def selection_sort(nums):
    for i in range(len(nums)):
        min_val_index = i
        for j in range(i + 1, len(nums)):
            if nums[j] < nums[min_val_index]:
                min_val_index = j
        nums[i], nums[min_val_index] = nums[min_val_index], nums[i]

random_list_of_nums = [0, 9, 5, 2, 1, 8, 4, 3, 7, 6]
selection_sort(random_list_of_nums)
print(random_list_of_nums)
```

## Сортироовка вставкой

```python
def insert_sort(nums):
    for i in range(1, len(nums)):
        x = nums[i]
        print('   start i =', i, nums, '\nx =', x)
        j = i - 1
        while j >= 0 and x < nums[j]:
            nums[j + 1] = nums[j]
            j -= 1
            print('\n', nums)
        nums[j + 1] = x
        print('\n', 'end i =', i, nums, '\n')
            
lst = [6, 9, 5, 2, 1, 8, 4, 3, 7, 6]
insert_sort(lst)
lst
```

## Сравнение скорости
Сравнение сортировки пузырьком, выбором и вставкой. Результаты для разного объема данных.

```python
import matplotlib.pyplot as plt

from datetime import datetime as dt
import numpy as np
start = [1, 10, 1000]
end = [11, 1001, 3001]
delta = [1, 50, 100]
repeat = 3
t_start = float(dt.utcnow().timestamp())

for j in range(repeat):

    # Сюда будем писать время сортировки
    lst_bubble = []
    lst_selection = []
    lst_insert = []

    # На каждом шаге генерируем случайный список длины i
    for i in range(start[j], end[j], delta[j]):
        l = list(np.random.rand(i))
        l2 = l.copy()
        l3 = l.copy()

        # Засекаем время и сортируем
        t0 = float(dt.utcnow().timestamp())
        bubble_sort(l)
        t1 = float(dt.utcnow().timestamp()) - t0
        lst_bubble.append(t1)


        t0 = float(dt.utcnow().timestamp())
        selection_sort(l2)
        t1 = float(dt.utcnow().timestamp()) - t0
        lst_selection.append(t1)


        t0 = float(dt.utcnow().timestamp())
        insert_sort(l3)
        t1 = float(dt.utcnow().timestamp()) - t0
        lst_insert.append(t1)
        
        

    # plot(список значений по оси X, список значений по оси Y, label = название линии на графике)
    plt.plot(range(start[j], end[j], delta[j]), lst_bubble)
    plt.plot(range(start[j], end[j], delta[j]), lst_selection)
    plt.plot(range(start[j], end[j], delta[j]), lst_insert)
    #plt.grid(True)
    plt.legend(['bubble','selection','insert'], loc=2)

    plt.show()
t_end = float(dt.utcnow().timestamp()) - t_start
print(f't_end = {t_end}')
```

## Сортироовка слиянием

```python
from random import randint

def merge(lst1, lst2):
    lst_res = []
    j = 0
    i = 0
    len_1 = len(lst1)
    len_2 = len(lst2)
    
    while i + j < len_1 + len_2:
        
        if len_1 > i and len_2 > j and lst1[i] <= lst2[j]:
            lst_res.append(lst1[i])
            i += 1
            
        elif len_1 > i and len_2 > j and lst1[i] > lst2[j]:
            lst_res.append(lst2[j])
            j += 1
            
        elif len_1 == i:
            while j < len_2:
                lst_res.append(lst2[j])
                j += 1
            break
        
        elif len_2 == j:
            while i < len_1:
                lst_res.append(lst1[i])
                i += 1
            break
    return lst_res

def merge_sort(nums):
    len_n = len(nums)
    if len_n <= 1:
        return nums
    lst1 = merge_sort(nums[:len_n // 2])
    lst2 = merge_sort(nums[len_n // 2:])
    return(merge(lst1, lst2))
    
    
for i in range(5):
    l = []
    for i in range(4):
        l.append(randint(0, 10))
    print('before', l, end='  ')
    print('after', merge_sort(l))
```

## Быстрая сортироовка
Опорный элемент выбирается случайно. Алгоритм не оптимизирован по памяти.

```python
from random import randint

def quick_sort(nums):
    ln = len(nums)
    if ln <= 1:
        return nums
    ind = randint(0, ln - 1)
    l_lst = []
    r_lst = []
    m_lst = []
    for i in range(ln):
        if nums[i] > nums[ind]:
            r_lst.append(nums[i])
        elif nums[i] < nums[ind]:
            l_lst.append(nums[i])
        elif nums[i] == nums[ind]:
            m_lst.append(nums[i])
    return quick_sort(l_lst) + m_lst + quick_sort(r_lst)


for i in range(5):
    l = []
    for i in range(4):
        l.append(randint(0, 10))
    print('before', l, end='  ')
    print('after', quick_sort(l))
```

## Быстрая сортироовка 2
        Опорный элемент в середине. Алгоритм оптимизирован по памяти. Возможна оптимизация по скорости.

```python
from random import randint


def part(nums, start, end):
    # x = randint(0, ln - 1)
    x = (end - start) // 2 + start
    a = nums[x]
    i_r = x + 1
    i_l = x - 1
    while i_r <= end:
        if nums[i_r] < a:
            if i_l >= start and nums[i_l] > a:
                nums[i_l], nums[i_r] = nums[i_r], nums[i_l]
            else:
                nums[x + 1], nums[i_r] = nums[i_r], nums[x + 1]
                nums[x], nums[x + 1] = nums[x + 1], nums[x]
                x += 1
            i_l -= 1
        i_r += 1
    
    while i_l >=  start:
        if nums[i_l] > a:
            nums[x - 1], nums[i_l] = nums[i_l], nums[x - 1]
            nums[x], nums[x - 1] = nums[x - 1], nums[x]
            x -= 1
        i_l -= 1
    return x
    
def quick_sort2(arr, start=0, end=None):
    if end == None:
        end = len(arr) - 1
    ln = end - start
    if ln < 1:
        return
    el = part(arr, start, end)
    quick_sort2(arr, el + 1, end)
    quick_sort2(arr, start, el - 1)
    return


for i in range(5):
    l = []
    for i in range(4):
        l.append(randint(0, 10))
    print('before', l, end='  ')
    quick_sort2(l)
    print('after', l)
```

## Быстрая сортироовка 3
        Опорный элемент в начале. Алгоритм оптимизирован по памяти и по скорости.

```python
# Функция принимает на вход список и индексы его начала и конца
def partition(array, start, end):
    #print(f'  start: arr[{start}] = {array[start]}, end = {end}, arr = {array}')
    # За опорный элемент выбрано начало (его индекс)
    pivot = start
    
    # Идем по списку от 1 элемента до конца
    # Как только встречаем элемент <= arr[start] меняем его местами 
    # с ближайшим к start элементом, который больше его(arr[pivot])
    # в конце меняем arr[start] и элемент < arr[start], который находится на max расстоянии от него
    for i in range(start+1, end+1):
        
        # Если элемент меньше или равен опорному
        if array[i] < array[start]:
            
            #Идем по списку от 1 элемента до конца
            pivot += 1
            array[i], array[pivot] = array[pivot], array[i]
        #print(f'i = {i}, pivot = {pivot}, arr = {array}')
    #print(f'pre end: pivot = {pivot}, start = {start}', array)
    array[pivot], array[start] = array[start], array[pivot]
    #print('end', array)
    return pivot


def quick_sort3(array, start=0, end=None):
    if end is None:
        end = len(array) - 1

    def _quicksort(array, start, end):
        if start >= end:
            return
        pivot = partition(array, start, end)
        _quicksort(array, start, pivot-1)
        _quicksort(array, pivot+1, end)
    return _quicksort(array, start, end)

array = [5, 6, 7, 2, 6, 0]
quick_sort3(array)
print(array)
```

```python
import matplotlib.pyplot as plt

from datetime import datetime as dt
import numpy as np
start = [1, 10, 100]
end = [11, 101, 4001]
delta = [1, 5, 100]
repeat = 3
t_start = float(dt.utcnow().timestamp())

for j in range(repeat):

    # Сюда будем писать время сортировки
    lst_merge = []
    lst_quick2 = []
    lst_quick = []
    lst_quick3 = []

    # На каждом шаге генерируем случайный список длины i
    for i in range(start[j], end[j], delta[j]):
        l = list(np.random.rand(i))
        l2 = l.copy()
        l3 = l.copy()
        l4 = l.copy()

        # Засекаем время и сортируем
        t0 = float(dt.utcnow().timestamp())
        merge_sort(l)
        t1 = float(dt.utcnow().timestamp()) - t0
        lst_merge.append(t1)


        t0 = float(dt.utcnow().timestamp())
        quick_sort2(l2)
        t1 = float(dt.utcnow().timestamp()) - t0
        lst_quick2.append(t1)


        t0 = float(dt.utcnow().timestamp())
        quick_sort(l3)
        t1 = float(dt.utcnow().timestamp()) - t0
        lst_quick.append(t1)
        
        t0 = float(dt.utcnow().timestamp())
        quick_sort3(l4)
        t1 = float(dt.utcnow().timestamp()) - t0
        lst_quick3.append(t1)
        

    # plot(список значений по оси X, список значений по оси Y, label = название линии на графике)
    plt.plot(range(start[j], end[j], delta[j]), lst_merge)
    plt.plot(range(start[j], end[j], delta[j]), lst_quick)
    plt.plot(range(start[j], end[j], delta[j]), lst_quick2)
    plt.plot(range(start[j], end[j], delta[j]), lst_quick3)
    #plt.grid(True)
    plt.legend(['merge', 'quick', 'quick2', 'quick3'], loc=2)

    plt.show()
t_end = float(dt.utcnow().timestamp()) - t_start
print(f't_end = {t_end}')
```
