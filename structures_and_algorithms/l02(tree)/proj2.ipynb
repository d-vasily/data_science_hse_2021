## Queue with priority

```python
class queue:
    def __init__(self, size=100):
        self.lim = size
        self.size = 0
        self.lst = [None] * (size + 1)
    
    
    def clear(self):
        self.lst = [None] * self.lim
        self.size = 0
        
        
    def swap(self, ind1, ind2):
        self.lst[ind1], self.lst[ind2] = self.lst[ind2], self.lst[ind1]
    
    
    def add(self, n):
        if self.size == self.lim:
            self.lst += [None] * self.size
            self.lim *= 2
        self.size += 1
        index = self.size
        self.lst[index] = n
        while(index > 1):
            parent_index = index // 2
            if self.lst[parent_index] > self.lst[index]:
                break
            self.lst[index], self.lst[parent_index] = self.lst[parent_index], self.lst[index]
            #self.swap(index, parent_index)
            index = parent_index
    
    
    def extract(self):
        if self.size == 0:
            return "CANNOT"
        ans = self.lst[1]
        self.lst[1], self.lst[self.size] = self.lst[self.size], self.lst[1]
        #self.swap(1, self.size)
        self.size -= 1
        index = 1
        while (index * 2 <= self.size):
            double_index = index * 2
            if double_index == self.size:
                if self.lst[index] < self.lst[double_index]:
                    self.lst[index], self.lst[double_index] = self.lst[double_index], self.lst[index]
                    #self.swap(index, double_index)
                return ans
            max_index = double_index
            if self.lst[double_index] < self.lst[double_index + 1]:
                max_index += 1
            if self.lst[index] >= self.lst[max_index]:
                return ans
            self.lst[index], self.lst[max_index] = self.lst[max_index], self.lst[index]
            #self.swap(index, max_index)
            index = max_index
        return ans
```

## Segment tree

```python
class Segment_tree:
    def __init__(self, lst):
        self.base = 1
        while (self.base < len(lst)):
            self.base *= 2
        self.lst = [None] * self.base * 2
        for i in range(len(lst)):
            self.lst[self.base + i] = lst[i]
        for i in range(self.base - 1, 0, -1):
            self.lst[i] = self.get_min(self.lst[i * 2], self.lst[i * 2 + 1])
    
    def get_min(self, elem_l, elem_r):
        if elem_r == None:
            return elem_l
        return min(elem_l, elem_r)        
        
    def __segment(self, pointer, search_l, search_r, seg_l, seg_r):
        if search_l == seg_l and search_r == seg_r:
            return self.lst[pointer]        
        new_border = (seg_l + seg_r) // 2
        if search_r <= new_border:
            return self.__segment(pointer * 2, search_l, search_r, seg_l, new_border)
        if search_l >= new_border + 1:
            return self.__segment(pointer * 2 + 1, search_l, search_r, new_border + 1, seg_r)
        l = self.__segment(pointer * 2, search_l, new_border, seg_l, new_border)
        r = self.__segment(pointer * 2 + 1, new_border + 1, search_r, new_border + 1, seg_r)
        return self.get_min(l, r)
    
    def get_min_seg(self, search_l, search_r):
        return self.__segment(1, search_l, search_r, 0, self.base - 1)
    
    
def test(n, k, lst):
    segment_tree = Segment_tree(lst)
    for i in range(n - k + 1):
        print(segment_tree.get_min_seg(i, i + k - 1))
```

```python
n, k = map(int, input().split())
lst = list(map(int, input().split()))
test(n, k, lst)
```

## Lagrange theorem

```python
from math import sqrt

def lagrange(m):
    n = m
    a, b, c, d = 0, 0, 0, 1
    while(a * a < n):
        b = a
        while(b * b < n):
            c = b
            while(c * c < n):
                d = c
                while(d * d < n):
                    sum = a * a + b * b + c * c + d * d
                    if sum == n:
                        #print(a, b, c, d, sum, end='\n\n')
                        return (a, b, c, d)
                    d += 1
                c += 1
            b += 1
        a += 1
    
def test_lagr(n):
    print(n)
    lagrange(n)

test_lagr(23)
```
